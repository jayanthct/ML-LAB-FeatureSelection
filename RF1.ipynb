```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

data = pd.read_csv('diabetes.csv')

data.head()

X = data.drop('Outcome', axis=1)
y = data['Outcome']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

forest_model = RandomForestClassifier(random_state=0)
forest_model.fit(X_train, y_train)

importance_values = forest_model.feature_importances_
importance_df = pd.DataFrame({'Attribute': X.columns, 'Value': importance_values})
importance_df = importance_df.sort_values(by='Value', ascending=False)

top_attributes = importance_df['Attribute'].head(5)

X_train_reduced = X_train[top_attributes]
X_test_reduced = X_test[top_attributes]
forest_reduced = RandomForestClassifier(random_state=0)
forest_reduced.fit(X_train_reduced, y_train)

predictions_reduced = forest_reduced.predict(X_test_reduced)

reduced_accuracy = accuracy_score(y_test, predictions_reduced)
print("Accuracy with reduced attributes:", reduced_accuracy)
```
