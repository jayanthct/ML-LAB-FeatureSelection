```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.feature_selection import RFE

data = pd.read_csv('diabetes.csv')

X = data.drop('Outcome', axis=1)
y = data['Outcome']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

tree_model = DecisionTreeClassifier()

feature_selector = RFE(tree_model, n_features_to_select=1, step=1)
feature_selector.fit(X_train, y_train)

chosen_features = X_train.columns[feature_selector.support_]

tree_model.fit(X_train[chosen_features], y_train)

y_predicted = tree_model.predict(X_test[chosen_features])

accuracy_val = accuracy_score(y_test, y_predicted)
print("Final set of features:", chosen_features)
print("Accuracy:", accuracy_val)
```
